# Notebook de Testes - ReAct Assistant
Demonstra todas as capacidades do agente

### Setup

```python
from react_assistant import ReActAssistant
import os

# Configure sua chave
os.environ["OPENAI_API_KEY"] = "sua_chave_aqui"

# Inicializa o agente
assistant = ReActAssistant()
```

### TESTE 1: Cálculo Simples

```python
print("=" * 80)
print("TESTE 1: Cálculo Matemático")
print("=" * 80)

result = assistant.run("Quanto é 15% de 2500?")
print(assistant.explain_reasoning(result))
```

### TESTE 2: Base de Conhecimento

```python
print("\n" + "=" * 80)
print("TESTE 2: Busca em Base de Conhecimento")
print("=" * 80)

result = assistant.run("O que é ReAct?")
print(assistant.explain_reasoning(result))
```
